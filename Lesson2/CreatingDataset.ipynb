## Classes for the classification

````python
classes = ['venomous','non_venomous']
````

### Taking the path of all the species as a list to download in a loop

````python
list1 = Path('VenomousSnakesURLs').ls()
dest = Path('data/venomous')
list2 = Path('NonVenomousSnakesURLs').ls()
dest2 = Path('data/non_venomous')
````

## Under vision.data
### download_images()
* Downloads images <br> param 1 : url or list <br> param 2 : destination path <br> param 3 : max pics to be dowloaded

### verify_images()
* try all the images in the path whether can it opened or not <br> param 1 : path of the dorectory to check <br> param 2 : bool to either to delete them if they are broken <br> param 3 : max_size to resize with the same ration but both sides less than the max size

````python
for i in list1:
    download_images(i, dest, max_pics=400)
````

````python
for i in list2:
    download_images(i, dest2, max_pics=400)
````

````python
path = Path('data')
for c in classes:
    verify_images(path/c, delete=True, max_size=500)
````

## Looking at the data and loading them to a databunch

````python
path = Path('data')
for c in classes:
    print(c, len(Path(path/c).ls()))
````

````python
np.random.seed(33)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2, ds_tfms=get_transforms(),
                                 size=299, num_workers=4, bs=32).normalize(imagenet_stats)
````

````python
data.show_batch(rows=3, figsize=(7,7))
````

## Simple training as this notebook's objective is dataset

````python
learn = create_cnn(data, models.resnet50, metrics=accuracy)
````

````python
learn.fit_one_cycle(2)
````

````python
interp = ClassificationInterpretation.from_learner(learn)
````

````python
interp.plot_confusion_matrix()
````

````python
interp.plot_top_losses(9)
````
